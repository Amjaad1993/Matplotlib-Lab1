```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
# Histogram
names = ["amjad", "jood", "ward"]
ages = [29, 3, 2]
data = {"names":names, "ages":ages}
df = pd.DataFrame(data)
plt.hist(df["ages"])
plt.xlabel('ages')
plt.ylabel('Frequency')
plt.title('Distribution of age')
plt.show()
```

```python
top_movies = df.groupby('ages').sum()
top_movies.head()
```

```python
# Bar Chart 
plt.bar(df['names'], df['ages'])
plt.xlabel('names',fontsize = 20)
plt.ylabel('ages',fontsize = 20)
plt.grid()
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python

plt.pie(df['ages'],labels=df['names'], autopct='%1.1f%%', startangle=90, shadow=True,
        radius=1, explode = [0,0,0])
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(df['names'],df['ages'], alpha = 0.4, s=50)
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.title('chart title', color= "red", loc="left")
plt.xlabel('names',fontsize = 16)
plt.ylabel('ages',fontsize = 16)
plt.show()
```

## Q5: Create three line charts with Legends at left and colors

```python
# Line chart

plt.plot([1,1,1,1,1])
plt.plot([2,2,2,2,2])
plt.plot([3,3,3,3,3])
plt.legend('legend', loc = 'upper left')
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=[20,10])
plt.suptitle('Subplots',fontsize = 20)

#first
plt.subplot(1,3,1) #(number of rows, number of columns, number of plot)
plt.pie(df['ages'],labels=df['names'], autopct='%1.1f%%')

#second
plt.subplot(1,3,2)
plt.scatter(df['ages'],df['names'])

#third
plt.subplot(1,3,3) #(number of rows, number of columns, number of plot)
plt.plot([1,1,1,1,1])
```
